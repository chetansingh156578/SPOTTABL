## Problem Statement 2

Given a domain (any one of the domains in ‘tags’ field), present the companies with the fastest, and the slowest growth (hint: it may be based on the fundingList, foundedDate, etc)

Let's assume the given tag, i'll go with 'pharmaceutical'.
So let's understand what do we infer from growth of a company. First let's print a few tuples of the dataset.

```python
df.head(5)
```

Here three coulumns are of my are of interest
- employeeNumber (Gives us an estimation of the size of a company in terms of manpower)
- foundedDate (type = dictionary)
- fundingList (type = list(dictionary))

I will here again get the info about the dataset

```python
df.info()
```

Here we can see that fundingList has a lot of null values present so it won't serve as a good metric to evaluate a company's growth so using the fundingList column is out of the picture.
However employee number and foundedDate provide a promising non-null value number so i will go ahead with using these two colums

I will here cleverely invent a company growth metric which will be defined as follows:-

**metric = (employeeNumber['max']-employeeNumber['min'])*(foundedDate(Year)+foundedDate(Month)+foundedYear(Date))/100000**

So what basically this metric will do is that it will us an estimate how much a company has expanded in size ever since it was founded. The **metric** variable will be of float type. Greater the value of metric greater the growth of the company and the opposite also holds true. 

So let's first get the location of the companies that have tag 'pharmaceutical'

```python
pharmaceutical = []
for i in range(len(list2)):
    if list2[i]:
        if str(type(list2[i])) == "<class 'float'>":
            pass
        else:
            if 'pharmaceutical' in list2[i]:
                pharmaceutical.append(i)
    
```

```python
pharmaceutical[:10]
```

So here we've got the companies with tags as pharmaceutical.
Let's also fetch the **employeeNumber** and **foundedDate** in the form of lists for better manipulation

```python
serie3, serie4 = df['employeeNumber'].iloc[pharmaceutical], df['foundedDate'].iloc[pharmaceutical]
```

```python
list3, list4 = list(serie3), list(serie4)
```

Now let's first get the information of our use from the list3 i.e employeeNumber(max)-employeeNumber(min)

```python
for i in range(len(list3)):
    if list3[i]:
        if str(type(list3[i])) == "<class 'float'>":
            pass
        else:
            numerator1 = list3[i]['max'] - list3[i]['min']
            list3[i] = numerator1
```

```python
list3[:10]
```

So here we have successfully got the first part of numerator values for our metrics

Now let's get the second part of the numerator values of our metrics

```python
for i in range(len(list4)):
    if list4[i]:
        if str(type(list4[i])) == "<class 'float'>":
            pass
        else:
            if str(type(list4[i]['$date'])) == "<class 'dict'>":  # in oder to handle numberLong dates
                pass
            else:
                temp = 0
                date = list4[i]['$date'][:10]
                date = date.split('-')
                temp += int(date[0])
                temp += int(date[1])
                temp += int(date[2])
                list4[i] = temp
            
        
```

```python
list4[:15]
```

So here we have obtained our numerator2 also, so let's calculate the metrics

```python
metrics = []

for i in range(len(list3)):
    if str(type(list3[i])) == "'<class 'float'>'" or str(type(list4[i])) == "'<class 'float'>'" or str(type(list4[i])) == "<class 'dict'>":
        metrics.append(-1)
    else:
        metrics.append((list3[i]*list4[i])/100000)
```

```python
metrics[:20]
```

So now we've got our metrics, now we can finally choose our top and least growing companies
So let's get our top ten pharmaceutical companies in terms of growth

```python
import math
```

```python
# Removing -1 and nan values from my dataset
temp = []
for i in metrics:
    if math.isnan(i) or i==-1:
        pass
    else:
        temp.append(i)
temp[:15]
```

```python
temp.sort(reverse=True)
```

```python
temp[:11]
```

```python
top10pharmaceutical = temp[:11]
```

```python
indexoftop10pharmaceutical = []

for i in range(len(metrics)):
    if metrics[i] == -1 or math.isnan(metrics[i]):
        pass
    elif metrics[i] in top10pharmaceutical:
        indexoftop10pharmaceutical.append(i)
        

indexoftop10pharmaceutical    
```

```python
ans1 = []
for i in indexoftop10pharmaceutical:
    ans1.append(pharmaceutical[i])
ans1
```

```python
df.iloc[ans1]
```

So here we've got our top ten pharmaceutical companies in terms of growth calculated based on my defined metric. Similarly we can get bottom ten companies in terms of growth calculated based on my defined metric.

```python
bottom10pharmaceutical = temp[-10:]
```

```python
bottom10pharmaceutical
```

```python
indexofbottom10pharmaceutical = []

for i in range(len(metrics)):
    if metrics[i] == -1 or math.isnan(metrics[i]):
        pass
    elif metrics[i] in bottom10pharmaceutical:
        indexofbottom10pharmaceutical.append(i)
        

indexofbottom10pharmaceutical   
```

```python
ans2 = []
for i in indexofbottom10pharmaceutical:
    ans2.append(pharmaceutical[i])
ans2
```

```python
df.iloc[ans1]
```

So here we've got bottom 10 pharmaceutical companies in terms of growth

**Note - The growth metric defined by me is based on the idea that the newer company with large company size in terms of manpower has good growth while an old company with small company manpower hasn't developed much. So the founded date of the company is in multiplication in numerator with the manpower of the comapny and 100000 in the denominator is used to scale the values to a much suitable range.**
